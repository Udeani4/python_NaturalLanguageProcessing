## STEMMING
Stemming is the process of reducing  word to its **word stem** that affixes to suffixes and prefixes or to the roots of words as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP). 

For example, we have a classification problem. We want to find out if the comments of product is a positive or negative review.

Words like eat, eating, eaten have a stem word as 'eat'. Those three words actually represent the same action, just different time instances. This is what stemming does for us



```python
words=["eating","eats","eaten","writing","writes","programming","programs","programmed","history","finally","finalized"]
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(f"{word}---->{stemming.stem(word)}")
```

You can observe some the words were not correctly stemmed, eg 'eaten' returned 'eaten'. Nevertheless the bigger issue is 'history' stemmed to 'histori' both words mean 2 different things. This is some of the disadvantages of stemming.

```python
## Another example

stemming.stem('congratulations')
```

See how the stem 'congratul' makes no sense and the meaning is not even close to the word 'congratulations'.

This issue can get fixed with the help of lemmatization. Although we can try other stemming techniques to improve it. We will try Regular Expression stemmer (**RegexpStemmer**)

### RegexpStemmer class
NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example...

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

Lets understand the regex argument provided.
`'ing$|s$|e$|able$', min=4`

This part `'ing$|s$|e$|able$'` means if the word ends with ing, s, e, or able, remove them. Therefore they are treated like suffixes. Hence why the dollar sign comes after the words

if the `'ing|s|e|able'` was the case, then It will remove them anywhere within the word.

Lets see some examples

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

You can see it only removes 'ing' at the back of the word

```python
reg_stemmer2=RegexpStemmer('ing|s$|e$|able$', min=4)
```

```python
reg_stemmer2.stem('ingeating')
```

Notce 'ing' in front and at the back got removed

```python
reg_stemmer2.stem('ingloringlywalking')
```

You observe that all 'ing' got removed, including the one within the word

### Snowball Stemmer

This one gives a better stemming accuracy than porter stemmer.

```python
from nltk.stem import SnowballStemmer
```

```python
snowball=SnowballStemmer('english')
```

```python
for word in words:
    print(word+'---->'+snowball.stem(word))
```

In this case the results are the same with that of Porter Stemmer. Lets see other examples where snowball is better.

```python
stemming
```

```python
snowball
```

You can see **snowball** identified the stem much better than **porter**.

Finally, we see that stemming is not very ideal because it can return some errorneous stems. We cannot apply techniques like these in development of chatbots. We have to use another technique called **lemmatization**

## LEMMATIZATION
Lemmatization technique is like stemming. The output we will get after lemmatization is called **'lemma'**, which is a root word rather than root stem( the output of stemming). After lemmatization, we will be getting a valid word that means the same thing.

### Wordnet Lemmatizer
NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example...

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going") ## The default pos is 'n' (noun). 
## This means the word you insert is taken as a noun
```

```python
lemmatizer.lemmatize("going",pos='v')
```

Notice when i changed the pos to 'v'(verb). It properly identified the root.

* 'a' - adjective
* 'r' - adverb
* 'n' - noun (default)
* 'v' - verb

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word))
```

Recall the default pos is 'n'. Now most of those words are verbs, so we will try pos='v')

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='v'))
```

Much better. So in lemmatization we are given the opportunity to decide what figure of speech the word is to improve its rooting accuracy. We can use it for:

* Q&A
* Chatbots
* Text summarization
