```python
corpus="""Hello welcome, to Udeani Izuchukwu's NLP Tutorials. 
Please do watch the entire course! 
To become expert is NLP."""
```

```python
print(corpus)
```

#### Convert Paragraph --> Words

```python
import nltk

packages = [
    "punkt",
    "punkt_tab",
    "stopwords",
    "wordnet",
    "omw-1.4",
    "averaged_perceptron_tagger"
]

for package in packages:
    nltk.download(package)
```

```python
from nltk.tokenize import sent_tokenize ## This will break the corpus(paragraph) to sentences
```

```python
documents=sent_tokenize(corpus)
documents
```

You can see the corpus is split into sentences and returned in a list.

#### Paragraph --> words OR Sentence --> words

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

You can see how every word is extracted from the paragraph or sentence. Notice punctuation marks like ".", ",", "!", "'s" were also considered as words. We can handle that when cleaning

```python
## Let us perform the word tokenization of the sentences

for sentence in documents:
    print(word_tokenize(sentence))
```

There are other libraries we can use as well eg **wordpunct_tokenize**

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

You can notice that the 's is now splitted. The wordpunct_tokenize ensures the apostrophes and punctuations are treated as words

```python
## TreebankWordTokenizer

from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```

With **TreebankWordTokenizer**, you can observe that fullstops are not treated as a separate word, **except** the last one at the end of the paragraph.
