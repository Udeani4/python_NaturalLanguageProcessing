# Converting Words to Vectors

After text preprocessing, we will now try to give meaning to each word for the computer to understand. This where we apply different techniques to convert each word to vectors. Some of the techniques used are:
* One Hot Encoding (OHE)
* Bag Of Words (BOW)
* Term Frequency-Inverse Document Frequency (TF-IDF)
* Word2Vec
* AvgWord2Vec

## BAG OF WORDS

```python
import pandas as pd
messages=pd.read_csv("SMSSpamCollection.txt",sep='\t',names=['label','message'])

## '\t' is a tab separated variable
```

```python
messages.head()
```

```python
## Data cleaning and preprocessing
import re ## import regular-expression
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    ## remove special characters
    review=re.sub('[^a-zA-z]', ' ', messages['message'][i]) ## This says- apart from small and capital a-z any other thing you find replace it with blankspace. Perform this is in the data provided
    review=review.lower()
    review_words=review.split()
    stemmed_review_words=[ps.stem(word) for word in review_words if not word in stopwords.words('english')]
    stemmed_review_sentence=' '.join(stemmed_review_words)
    corpus.append(stemmed_review_sentence)
```

```python
corpus
```

```python
## Create the bag of words (BOW) model
## We will use sklearn's CountVectorizer for this
from sklearn.feature_extraction.text import CountVectorizer
```

```python
countvectorizer=CountVectorizer(max_features=2500)
## max_features=2500 is saying that out of all the words in the BOW take the top 2500 words
## for binary BOW set binary=True
```

```python
X=countvectorizer.fit_transform(corpus).toarray()
```

## SPAM HAM PROJECT

Here we will develop a model to predict if a message is a spam or not. We will apply `BOW`, `TF-IDF`, `Word2Vec` and `AvgWord2Vec` (for Vectorizing) and `Machine Learning Algorithms` for training

```python
import pandas as pd
messages = pd.read_csv('SMSSpamCollection.txt',sep='\t',names=['label','message'])
```

```python
messages
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
sbs=SnowballStemmer('english')
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[sbs.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

Now we are going to build our own word2vec from scratch with the spam dataset.

```python
import pandas as pd
messages=pd.read_csv("SMSSpamCollection.txt", sep='\t', names=['label','message'])
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()

    # review=[lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    ## we will not filter out the stop words here. Let the lemmatizer do its job
    review=[lemmatizer.lemmatize(word) for word in review] ## The lemmatizer will stem/root the words including stop words
    review=' '.join(review)
    corpus.append(review)
```

Lets check out other processes to handle what we have just don above

```python
from nltk import sent_tokenize ## Tokenize paragraphs to sentences
from gensim.utils import simple_preprocess ## Convert a document into a list of lowercase tokens, ignoring tokens that are too short or too long.
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
## Lets train word2vec from scratch. Rather than using google pretrained word2vec model
import gensim
w2v_model=gensim.models.Word2Vec(
    sentences=words,
    vector_size=100, ## recall google used 300. 100 is default
)
```

Now we can see that for every word we have created a 100 dimension vector. What we need is a vector for each sentence. That is where **`AvgWord2Vec`** comes in. We will calculate the average of the vectors for each word in the sentence to get a vector representation for the sentence.

PS: This model has not yet been improved. This is just practicals.

```python
import numpy as np

# Krish func
# def avg_word2vec(doc):
#     ## remove out-of-vocabulary words
#     #sent=[word for word.in doc if word is in w2v_model.wv.index_to_key]
#     #print(sent)

#     return np.mean([w2v_model.wv[word] for word in doc if word in w2v_model.wv.index_to_key],axis=0)

# Izu func. Created because a sentence can have no word in the model vocabulary and then it will return an empty list
# I captured that and returned zeros vector instead
def avg_word2vec(doc):
    vectors = [
        w2v_model.wv[word]
        for word in doc
        if word in w2v_model.wv
    ]

    if len(vectors) == 0:
        return np.zeros(w2v_model.vector_size)

    return np.mean(vectors, axis=0)
```
