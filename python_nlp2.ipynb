* Bag Of Words (BOW)

## BAG OF WORDS

```python
import pandas as pd
messages=pd.read_csv("SMSSpamCollection.txt",sep='\t',names=['label','message'])

## '\t' is a tab separated variable
```

```python
messages.head()
```

```python
## Data cleaning and preprocessing
import re ## import regular-expression
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    ## remove special characters
    review=re.sub('[^a-zA-z]', ' ', messages['message'][i]) ## This says- apart from small and capital a-z any other thing you find replace it with blankspace. Perform this is in the data provided
    review=review.lower()
    review_words=review.split()
    stemmed_review_words=[ps.stem(word) for word in review_words if not word in stopwords.words('english')]
    stemmed_review_sentence=' '.join(stemmed_review_words)
    corpus.append(stemmed_review_sentence)
```

```python
corpus
```

```python
## Create the bag of words (BOW) model
## We will use sklearn's CountVectorizer for this
from sklearn.feature_extraction.text import CountVectorizer
```

```python
countvectorizer=CountVectorizer(max_features=2500)
## max_features=2500 is saying that out of all the words in the BOW take the top 2500 words
## for binary BOW set binary=True
```

```python
X=countvectorizer.fit_transform(corpus).toarray()
```

```python
X
```

```python
X.shape
```

### N Grams

We use this to create combination of words for improved sentence to vector transformation. Examples are unigram, bigrams, trigram etc

We will work an exampleon how to use N-grams

```python
## Create bag of words model
from sklearn.feature_extraction.text import CountVectorizer
## for binary BOW
cv=CountVectorizer(max_features=100,binary=True)
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(
    edgeitems=30,
    linewidth=100000,
    formatter=dict(float=lambda x: "%.3g" % x))

X
```

```python
cv.vocabulary_ ## These re the top 100 vocabularies with their column index
```

```python
len(cv.vocabulary_)
```

```python
## Now let us implement n-gram
## Create the Bag Of Words model with ngram
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=100,binary=True,ngram_range=(1,1)) 
## with ngram_range as (1,1) the values will be tha same as the previous. This is unigram
X=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
## Create the Bag Of Words model with ngram
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=200,binary=True,ngram_range=(1,2)) 
## max features increased to 200 for more illustration
## with ngram_range as (1,2). Here we combine unigram and bigram
X=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

If you go through the dictionary you will find some double word combination. Please refer to note for detailed explanation

```python
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=200,binary=True,ngram_range=(2,2)) 
## max features increased to 200 for more illustration
## with ngram_range as (2,2). Here we combine using bigram only (ignoring unigram)
X=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

You can do the same for trigram (3,3). Note:(1,3) will run unigram, bigram and trigram. When training you can run hyperparameters to test different combinations to get the best accuracy.

## TEXT FREQUENCY-INVERSE DOCUMENT FREQUENCY (TF-IDF)

Check note for explanation.

We will carry out the same process we carried out before.

```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection.txt',sep='\t',names=['label','message'])
messages
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    ## remove special characters
    review=re.sub('[^a-zA-z]', ' ', messages['message'][i]) ## This says- apart from small and capital a-z any other thing you find replace it with blankspace. Perform this is in the data provided
    review=review.lower()
    review_words=review.split()
    stemmed_review_words=[wnl.lemmatize(word) for word in review_words if not word in stopwords.words('english')]
    stemmed_review_sentence=' '.join(stemmed_review_words)
    corpus.append(stemmed_review_sentence)
```

```python
corpus
```

```python
## Create TF-IDF and NGrams
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=100)
X=tfidf.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(
    edgeitems=30,
    linewidth=100000,
    formatter=dict(float=lambda x: "%.3g" % x)
)
```

## SPAM HAM PROJECT

Here we will develop a model to predict if a message is a spam or not. We will apply `BOW`, `TD-IDF` and `Machine Learning Algorithms`

```python
import pandas as pd
messages = pd.read_csv('SMSSpamCollection.txt',sep='\t',names=['label','message'])
```

```python
messages
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
sbs=SnowballStemmer('english')
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[sbs.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## CREATE BAG OF WORDS MODEL
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=100)
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(
    edgeitems=30,
    linewidth=100000,
    formatter=dict(float=lambda x: "%.3g" % x))

X
```
